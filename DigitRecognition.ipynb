```python
# Import Necessary Libraries:
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize pixel values

# Reshape the data to (28, 28, 1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Flatten the 28x28 input images
    Dense(128, activation='relu'),
    Dropout(0.2),  # Optional dropout layer for regularization
    Dense(10, activation='softmax')  # 10 output units for 0-9 digits
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5)  # You can adjust the number of epochs
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_accuracy}")
```

```python
model.save('your_model.h5')  # Replace 'your_model' with your desired filename
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

# Load and preprocess the image
img_path = 'd.png'  # Replace with your image file
img = image.load_img(img_path, color_mode='grayscale', target_size=(28, 28))
img = image.img_to_array(img)
img = img / 255.0  # Normalize pixel values

# Reshape the image to match the model's input shape
img = np.expand_dims(img, axis=0)

# Make predictions
predictions = model.predict(img)

# Get the recognized digit
r = np.argmax(predictions)

print(f"Recognized Digit: {r}")
```
