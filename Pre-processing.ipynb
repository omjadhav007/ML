```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

# Load the MNIST dataset
mnist = fetch_openml("mnist_784")
X = mnist.data
y = mnist.target.astype(int)

# Select a subset of features (pixels)
n_samples = 1000  # Number of samples to visualize
n_features = 100  # Number of features (pixels) to visualize

# Select a random subset of samples and features
subset_indices = np.random.choice(X.shape[0], n_samples, replace=False)
X_subset = X[subset_indices, :n_features]

# Create a DataFrame for the selected features
import pandas as pd
df = pd.DataFrame(X_subset, columns=[f"Pixel_{i}" for i in range(n_features)])

# Add the target (label) to the DataFrame
df['Digit'] = y[subset_indices]

# Create a pairplot
sns.pairplot(df, hue='Digit', markers='o', plot_kws={'alpha': 0.5})
plt.show()
```
