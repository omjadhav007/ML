```python
# Roll No : RBT21CB055
# Problem Statement : State wheater preprocessing required or not for any discrete dataset.Perform preprocessing if required and apply classification on that discrete dataset.

# Importing required laibraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
```

```python
# Here initially I am using continuous dataset Iris
# So Multinomial Naive Bayes will give less accuracy
dataset = pd.read_csv('iris.csv')
dataset.head()
```

```python
x = dataset.drop(columns=['species'])
y = dataset['species']
x.head()
```

```python
y.head()
```

```python
xTr,xTst,yTr,yTst=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
klacifier = MultinomialNB()
klacifier.fit(xTr,yTr)
prredict = klacifier.predict(xTst)
accur = accuracy_score(yTst,prredict)
print(f"Accuracy of Multinomial Naive Bayes : {accur}")
```

```python
c = GaussianNB()
c.fit(xTr,yTr)
pr = c.predict(xTst)
ac = accuracy_score(yTst,pr)
print(f"Accuracy of Gaussian Naive Bayes : {ac}")
```

```python
# Here, Accuracy of Gaussian Naive Bayes > Accuracy of Multinomial Naive Bayes.
```

```python
# Now I am using discrete dataset Email-spam.
# So Multinomial Naive Bayes will give more accuracy than Gaussian Naive Bayes

dataset = pd.read_csv('email-spam.csv')
dataset.head()
```

```python
x = dataset.drop(columns=['class'])
y = dataset['class']
x.head()
```

```python
y.head()
```

```python
xTr,xTst,yTr,yTst=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
klacifier = MultinomialNB()
klacifier.fit(xTr,yTr)
prredict = klacifier.predict(xTst)
accur = accuracy_score(yTst,prredict)
print(f"Accuracy of Multinomial Naive Bayes : {accur}")
```

```python
c = GaussianNB()
c.fit(xTr,yTr)
pr = c.predict(xTst)
ac = accuracy_score(yTst,pr)
print(f"Accuracy of Gaussian Naive Bayes : {ac}")
```

```python
# Here, Accuracy of Gaussian Naive Bayes < Accuracy of Multinomial Naive Bayes
```

```python
c = BernoulliNB()
c.fit(xTr,yTr)
pr = c.predict(xTst)
ac = accuracy_score(yTst,pr)
print(f"Accuracy of Bernoulli Naive Bayes : {ac}")
```

```python
# In Email-spam dataset, classes are binary.
# So we can use Bernoulli Naive Bayes.
# Bernoulli Naive Bayes is giving accuracy approximately equal to Accuracy of Multinomial Naive Bayes.
```
